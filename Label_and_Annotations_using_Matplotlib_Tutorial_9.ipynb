# Labels & Annotations Overview 


- Labeling plot features

- Adding a legend to your plot

- Annotating you plot 

- Functional Method

- Object - oriented Method

# Annotating Your Plot

- To annotate some features of the plot
  - annotate(xy,xytext,and arrowprop)


- Parameters
  - xy = the location being annotaged
  - xytext = the location of the text
    - arrowprop = draws an arrow from the text to the annotated point by giving a dictionary of arrow properties

# Adding a Legend


- placing a legend on plot axes
  - legend(label & loc)

- parameters
  - label = variable or category labels
  - loc = the location of the legend

# **Creating labels and annotations**

```
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt

%matplotlib inline

from pylab import rcParams
import seaborn as sns
```

```
rcParams['figure.figsize']= 5,4
sns.set_style('whitegrid')
```

# **Labeling plot features**

## The functional method

```
x = range(1,10)
y = [1,2,3,4,0,4,3,2,1]

plt.bar(x,y)

plt.xlabel('your x-axis label')
plt.ylabel('your y-axis label')
```

```
z =[1,2,3,4,5]
veh_type = ['bicycle','motorbike','car','van','stroller']

plt.pie(z,labels= veh_type)

plt.show()
```

# The object-oriented method

```
from google.colab import drive
drive.mount('/content/drive')
```

```
cars = pd.read_csv('/content/drive/My Drive/MyData/mtcars.csv')
cars.head()
```

```
cars.columns =['car_names', 'mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs',
       'am', 'gear', 'carb']
     
```

```
mpg = cars.mpg

fig = plt.figure()

ax = fig.add_axes([.1,.1,1,1])

mpg.plot()
ax.set_xticks(range(32))

ax.set_xticklabels(cars.car_names,rotation= 60,fontsize= 'medium')
ax.set_title('MIles per Gallon of Cars in Mtcars')

ax.set_xlabel('car names')
ax.set_ylabel('miles/gal')
```

# **Adding a legend to your plot**

## The functional method

```
plt.pie(z)

plt.legend(veh_type,loc= 'best')

plt.show()
```

## The object-oriented method

```
fig = plt.figure()

ax = fig.add_axes([.1,.1,1,1])
mpg.plot()

ax.set_xticklabels(cars.car_names,rotation= 60,fontsize= 'medium')
ax.set_title('MIles per Gallon of Cars in Mtcars')

ax.set_xlabel('car names')
ax.set_ylabel('miles/gal')

ax.legend(loc= 'best')
```

```
fig = plt.figure()

ax = fig.add_axes([.1,.1,1,1])
mpg.plot()

ax.set_title('MIles per Gallon of Cars in Mtcars')

ax.set_ylabel('miles/gal')
ax.set_ylim([0,45])

ax.annotate('Toyota Corolla',xy =(19,33.9),xytext = (21,35),
            arrowprops = dict(facecolor= 'black',shrink= 0.05))
```
