```
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('retina')
```

```
plt.figure(figsize=(15,4))
plt.xticks(rotation= 60)
sns.pointplot(data=df,x='University',y='Graduation rate')
```

```
plt.figure(figsize=(15,4))
plt.xticks(rotation= 60)
sns.lineplot(data=df,x='University',y='Graduation rate')
```

```
df_pop = df[['University',  'Graduation rate', '\nClass rate of 20 or less','Number of students per professor',
             '\nAlumni donation rate']]
df_pop = df_pop.set_index('University')
df_pop.head()
```

```
plt.figure(figsize=(10,8))
sns.scatterplot(data=df_pop,x='Graduation rate',y='\nClass rate of 20 or less',hue='Number of students per professor')
```

```
df_pop.describe()
```

```
sns.lmplot(x='Graduation rate',y='\nClass rate of 20 or less',data=df_pop)
```

```
sns.lmplot(x='Graduation rate',y='Number of students per professor',data=df_pop)
```

```
sns.lmplot(x='Graduation rate',y='\nAlumni donation rate',data=df_pop)
```

```
plt.figure(figsize=(20,4))
sns.barplot(data=df,x='University', y='Graduation rate', palette="rocket").axhline(0, color="k", clip_on=False,)
```

```
plt.figure(figsize=(20,4))
sns.barplot(data=df,x='Graduation rate', y='Number of students per professor', palette="deep").axhline(0, color="k", clip_on=False)
```
