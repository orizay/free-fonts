# 연습

### 30대 여성고개데이터만 추출해서 구매금액이 높은 순서대로 파일을 만들어서 제공해주세요.

* 30대 여성 추출 -->> 구매금액 높은순서로 정렬
* 파일로 저장

```python
cond3 = (df1['연령대'] == '30대이하')
cond4 = (df1['성별'] == '여')

df1[cond3 & cond4]
```

```python
df3 = df1[cond3 & cond4].sort_values(by='구매금액',ascending=False)
```

```python
df3
```

```python
df3.shape
```
