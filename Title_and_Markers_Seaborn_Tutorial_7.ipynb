# Two Methods for Plot Building


- **Functional Method:**
   - with the functional method,you build plots by calling the plotting function on a variable or set of variables.


- **Object -Oriented Method:**
  - with the object-oriented method,you build a plot by first generating a blank 'figure'object,and then populanting that object with plots and plot elements.(Covered after this video portion)

# Steps in Object - Oriented Plotting


1.Create a blank figure object


2.Add axes to the figure


3.Generate plot(s)within the figure object

4.Specify plotting and layout parameters for plots within the figure

# Generating Subplots

![대체 텍스트](https://matplotlib.org/3.1.0/_images/sphx_glr_subplots_demo_005.png)


- A subplot is a plotting figure that contains more than on plot - or subplots.

- It's easy to generate a subplot in matplotlib using subplots().

# **Defining Elements of a Plot**

```
import numpy as np
from numpy.random import randn
import pandas as pd
from pandas import Series,DataFrame

import matplotlib.pyplot as plt
from matplotlib import rcParams

%matplotlib inline
```

```
rcParams['figure.figsize'] = 5,4
```

# **Defining axes,ticks,and grids**

```
x = range(1,10)
y = [1,2,3,4,5,0,4,3,2]

fig = plt.figure()

ax = fig.add_axes([.1,.1,1,1])

ax.plot(x,y)
```

```
fig = plt.figure()

ax = fig.add_axes([.1,.1,1,1])

ax.set_xlim([1,9])
ax.set_ylim([0,5])

ax.set_xticks([0,1,2,4,5,6,8,9,10])
ax.set_yticks([0,1,2,3,4,5])

ax.plot(x,y)
```

```
fig = plt.figure()

ax = fig.add_axes([.1,.1,1,1])

ax.set_xlim([1,9])
ax.set_ylim([0,5])

ax.grid()

ax.plot(x,y)
```

# **Generating multiple plots in one figure with subplots**

```
fig = plt.figure()

fig,(ax1,ax2) = plt.subplots(1,2)

ax1.plot(x)

ax2.plot(x,y)
```
